```python
df['age_category']=pd.cut(df['age'],bins=[0,2,15,25,60,100],labels=['infants','child','teen','adult','senior'],right=True,include_lowest=True)
```

```python
def companion_type(row):
    if row['Passenger Count']==1:
        'solo'
    elif row['FAMILY_COUNT']==0 and row['Passenger Count']>1:
        return 'friends'
    else:
        return 'family'
df['companion type']=df.apply(companion_type,axis=1)
    
```
